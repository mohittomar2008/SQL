```python
from pandasql import sqldf
```

```python
pysqldf= lambda q: sqldf(q,globals())
```

```python
import pandas as pd
```

```python
df=pd.read_csv('titanic_train.csv')
```

```python
df.head()
```

```python
df.drop(df.columns[0], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.columns[0]
```

```python
df.head()
```

```python
# lets slice DF by fare

fares_5_100 =df[(df.Fare >50)&(df.Fare <100)   ]
```

```python
fares_5_100.head()
```

```python
# For using SQL query

q= "select * from df where (Fare >50) & (Fare <100)"

fare50_100= pysqldf(q)
```

```python
fare50_100.head()
```

```python
# compare 2 results with .equal().  But notice sql query index got reset so we set the index with passengerid then use equal()
```

```python
fare50_100.set_index('PassengerId').equals(fares_5_100.set_index('PassengerId'))
```

```python
# 2nd example, get first 30 names of male passenger who survived
```

```python
# with python:

surviving_male_py= df[(df.Sex=='male')&(df.Survived==1) ][['Name','Fare']][:30]
```

```python
surviving_male_py
```

```python
q2="select * from df where (Sex='male') &(Survived=1 ) limit 30"
```

```python
surviving_male= pysqldf(q2)
```

```python
surviving_male
```

```python
# compare both

surviving_male.set_index('Name').equals(surviving_male_py.set_index('Name')  )
```
